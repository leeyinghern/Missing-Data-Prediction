```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("https://gist.githubusercontent.com/tmramalho/51733432c88e2b834dbd70353849f887/raw/676636d9e8512963005a318cb9b98060feadb641/data.csv")

df.head()
```

```python
df['x'] = df['x'].apply(lambda x: np.nan if x == "-" else x)
df['y'] = df['y'].apply(lambda x: np.nan if x == "-" else x)
```

```python
df.x = df.x.astype('float32')
df.y = df.y.astype('float32')
df
```

```python
plt.figure(figsize=(20,10))
plt.plot(df['t'],df['x'], label="Prey")
plt.plot(df['t'],df['y'], label="Predator")
plt.legend();
```

```python
df.dropna(inplace=True,thresh=3)
df
```

```python
HORIZON = 1
WINDOW_SIZE = 7
```

```python
train = tf.keras.utils.timeseries_dataset_from_array(df[:int(0.8*len(df))]['t'], 
                                             df[:int(0.8*len(df))][['x','y']][WINDOW_SIZE::WINDOW_SIZE],
                                             sequence_length = WINDOW_SIZE,
                                            )
test = tf.keras.utils.timeseries_dataset_from_array(df[int(0.8*len(df)):]['t'], 
                                             df[int(0.8*len(df)):][['x','y']][WINDOW_SIZE::WINDOW_SIZE],
                                             sequence_length = WINDOW_SIZE,
                                            )
```

```python
x_train = []
y_train = []
x_test = []
y_test = []

for batch in train:
    inputs, targets = batch
    x_train.append(inputs.numpy().tolist())
    y_train.append(targets.numpy().tolist())

for batch in test:
    inputs, targets = batch
    x_test.append(inputs.numpy().tolist())
    y_test.append(targets.numpy().tolist())
```

```python
train = pd.DataFrame(x_train).T
train['sin'] = train[0].apply(lambda x: [np.sin(i) for i in x])
train['combined'] = train[0]+ train['sin']


test = pd.DataFrame(x_test).T
test['sin'] = test[0].apply(lambda x: [np.sin(i) for i in x])
test['combined'] = test[0]+ test['sin']
```

```python
import os

def create_model_checkpoint(model_name, save_path="model_experiments"):
    return tf.keras.callbacks.ModelCheckpoint(filepath = os.path.join(save_path, model_name),
                                              verbose = 0,
                                              save_best_only=True
                                             )
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))
```

```python
from tensorflow.keras import layers
tf.random.set_seed(42)

model = tf.keras.Sequential()
model.add(layers.Dense(150, activation = "relu"))
model.add(layers.Dense(50, activation = "relu"))
model.add(layers.Dense(2, activation="linear"))

model.compile(loss=tf.keras.losses.mse,
              optimizer=tf.keras.optimizers.Adam(),
              metrics=[tf.keras.losses.mae,tf.keras.losses.mse]
             )

history = model.fit([[x for x in train.combined]], 
          y_train, 
          epochs = 500, 
          validation_data = ([[x for x in test.combined]], y_test),
          callbacks = [create_model_checkpoint("model_1"),lr_scheduler],
          batch_size = 32
         )
```

```python
perf = pd.DataFrame(history.history)
perf.loc[perf["val_loss"]==min(perf["val_loss"])]
```

```python
from tensorflow.keras import layers
tf.random.set_seed(42)

model = tf.keras.Sequential()
model.add(layers.Dense(150, activation = "relu"))
model.add(layers.Dense(50, activation = "relu"))
model.add(layers.Dense(20, activation = "relu"))
model.add(layers.Dense(2, activation="linear"))

model.compile(loss=tf.keras.losses.mse,
              optimizer=tf.keras.optimizers.Adam(learning_rate = 0.000282),
              metrics=[tf.keras.losses.mae,tf.keras.losses.mse]
             )

history = model.fit([[x for x in train.combined]], 
          y_train, 
          epochs = 500, 
          validation_data = ([[x for x in test.combined]], y_test),
          callbacks = [create_model_checkpoint("model_1")],
          batch_size = 32
         )
```

```python
train_pred = model.predict([[x for x in train.combined]])
test_pred = model.predict([[x for x in test.combined]])
```
