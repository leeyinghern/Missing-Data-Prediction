### Time series

<ul>
    <li>Ordered inputs that are spaced equally over periods.</li>
    <li>Data can be univariate or multivariate meaning that it can have one or more than one feature</li>
</ul>

<ul>
    <b> Common patterns in time series </b>
    <li>Trend: When a time series has a specific direction that it is moving in</li>
    <li>Seasonality: When patterns repeat at predictable intervals. (i.e. when time series is affected by seasonal factors like time of the year)</li>
    <li>Cyclicity: Time series that rises and falls over an unfixed period. These tend to be more variable than seasonal patterns</li>
    <li>White noise: Neither trend or seasonality</li>
    <li>Autocorrelation: When the next value correlates or dependednt with a delayed copy of itself at a lag. A time series like this is called as having memory.</li>
    <li>Time series usually consists of trend, seasonality, noise and autocorrelation</li>
</ul>

<b>Stationary time series: Time series that has behaviour that does not change over time.</b> For stationary data, we should use more data(the whole dataset) to train our model.<br><br>
<b>Non-Stationary time series: When the time series experiences a change in behavior that breaks trend and seasonality</b>. For example, a time series that previously had updwards trend and seasonality, but suddenly dropped to not follow any trend or seasonality is said to be non-stationary. For non-stationary data, we can just train on limited data. The optimal window we use to train the data will vary<br><br>

### Measuring Performance

<h4>Train-Validation-Test split</h4>
<ul>
    <li><b>Fixed Partitioning:</b>To measure performance for a time series, we split the data into a training period, a validation period and a test period. If the time series contains some sort of seasonality, we want to ensure that each period has a whole number of seasons. For example, one year, two or three years if the time series has yearly seasonality. The validation data is used to evaluate the model during training. Once satisfied, train the model on training and validation data. The test data is used to access the generalization ability of the model.</li>
    <li><b>Roll forward partitioning:</b> We start with a short training period and gradually increase it, say by one day or one week at a time. At each iteration, we train the model and use it to forecast the following day or week in the validation period.</li>
</ul>

<h4>Error Metrics</h4>
<ul>
    <li>MSE: Most common metric to measure prediction error. Squared to get rid of negative errors and change it to positive. <b>Use when large errors are more significant than smaller errors</b></li>
    <li>RMSE: To get the unit of error back to the original</li>
    <li>MAE: Does not penalize large errors like the MSE does.</li>
    <b>Choice between MSE and MAE depends on the task at hand. If large errors are more costly than smaller errors, then maybe MSE is more worthwhile. If the gain or loss is just proportional to the size of the error, then we may prefer MAE.</b>
    <li>MAPE: Mean absolute percentage error. Mean ratio between the absolute error and the absolute value. This gives an idea of the size of the errors compared to the values.</li>
    <li>MASE: Mean absolute percentage error. A scaled error > 1 if the forecast is worse than the naive and < 1 if the forecast is better than the naive</li>
</ul>
<h4>For all of these metrics, the lower the better</h4> <br><br>

## Moving average and Differencing (Statistical Forecasts)

<ul>
    <b>Moving Average</b>
    <li>Simple forecasting method. Yellow line = average of the blue values over a fixed period of time (window size)</li>
    <li>Eliminates noise but does not anticipate trend or seasonality</li>
</ul>
<div>
    
</div>

<ul>
    <b>Naive forecast</b>
    <li>Naive forecasts take the value at the previous timestep and sets it as the predicted value for the next timestep</li>
</ul>

<ul>
    <b>Differencing</b>
    <li>Removes trend and seasonality. We subtract each value of the time series at time t with it's corresponding value at time t - period. The period that we subtract depends on the data that we have. (i.e. if the seasonality is yearly seasonality, we have t - (t - 365) )</li>
    <li><b>When we difference, we are left with data + noise</b></li>
    <li>Calculate the moving average between the differenced points. Then we add the differenced moving average + the past values at time (t - period) to add back the trend + seasonality.</li>
    <li> We can smooth out the forecasts by adding the differenced moving average of the differenced series + the moving average of the PAST series at time (t - period). This removes the noise of the previous time period</li>
</ul>
